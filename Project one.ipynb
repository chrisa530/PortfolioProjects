```python
import pandas as pd
import matplotlib.pyplot as plt

# Select rows with School DBN '22K425' and '26Q430'
selected_rows = df.loc[df['School DBN'].isin(['22K425', '26Q430'])]

# Filter rows with School Name 'James Madison High School' and 'Francis Lewis High School'
selected_rows = selected_rows.loc[selected_rows['School Name'].isin(['James Madison High School', 'Francis Lewis High School'])]

# Filter rows with 'Physical Settings/Earth Science' in 'Regents Exam'
selected_rows = selected_rows.loc[selected_rows['Regents Exam'] == 'Physical Settings/Earth Science']

# Remove any rows with null values
selected_rows = selected_rows.dropna()

# Convert 'Number Scoring Below 65' column to numeric
selected_rows['Number Scoring Below 65'] = pd.to_numeric(selected_rows['Number Scoring Below 65'], errors='coerce')

# Sum the values for each school
sum_values = selected_rows.groupby('School Name')['Number Scoring Below 65'].sum()

# Calculate the total count of students for each school
total_count = selected_rows.groupby('School Name')['Total Tested'].sum()

# Calculate the percentage of scores below 65% for each school
percentage_below_65 = (sum_values / total_count) * 100

# Create a bar graph of the percentages
fig, ax = plt.subplots(figsize=(8, 6))  # Adjust the figure size
percentage_below_65.plot(kind='bar', ax=ax, color=['#FF6F61', '#6B5B95'])  # Customize the bar color

ax.set_xlabel('School Name', fontsize=12)  # Set the x-axis label and font size
ax.set_ylabel('Percentage Below 65%', fontsize=12)  # Set the y-axis label and font size
ax.set_title('Percentage of Scores Below 65%\nPhysical Settings/Earth Science', fontsize=14, fontweight='bold')  # Set the title and font size

ax.spines['top'].set_visible(False)  # Remove the top border
ax.spines['right'].set_visible(False)  # Remove the right border

plt.xticks(rotation=0)  # Rotate the x-axis labels if needed

plt.tight_layout()  # Adjust the layout to prevent label overlapping

plt.show()
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/2014-15_To_2016-17_School-_Level_NYC_Regents_Report_For_All_Variables.csv')
```

```python
import pandas as pd

# Assuming you have a DataFrame named 'df' containing the data

# Convert the "year" column to object (string) data type
df['Year'] = df['Year'].astype(str)

# Filter out rows containing 's' or 'na' in any of the columns
df = df[~df.apply(lambda row: any('s' in str(val) or 'na' in str(val) for val in row), axis=1)]

#Use https://www.geeksforgeeks.org/lambda-filter-python-examples/ 

# Convert the specified columns to numeric data type (numeric values stored as strings)
numeric_columns = ['Mean Score', 'Number Scoring Below 65', 'Number Scoring 65 or Above',
                   'Number Scoring 80 or Above', 'Number Scoring CR']

df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')

# Drop rows with any NaN values in any column
df.dropna(inplace=True)
```

## Code explanation help
### df = df[~df.apply(lambda row: any('s' in str(val) or 'na' in str(val) for val in row), axis=1)]


df.apply(...): The apply() method in Pandas is used to apply a function along the axis of the DataFrame. In this case, we are applying a lambda function to each row of the DataFrame.

lambda row: ...: We are defining a lambda function that takes a row as input and performs some operations on that row.

for val in row: The lambda function iterates over each value in the row.

's' in str(val): We are checking if the string representation of the current value (val) contains the character 's'. This is done to identify any values with the letter 's' in them, which might indicate an invalid or non-numeric value.

'na' in str(val): Similarly, we are checking if the string representation of the current value (val) contains the characters 'na'. This is done to identify any values that might be 'na' or 'nan', indicating missing or invalid data.

any(...): The any() function is a built-in Python function that returns True if at least one element in the given iterable is True. In this case, it checks if any of the conditions ('s' in val or 'na' in val) are true for any value in the row.

df.apply(...): The outer apply() function applies the lambda function to each row of the DataFrame.

~: The tilde (~) operator is a bitwise NOT operator. In this context, it is used to negate the boolean values returned by the apply() function. This means that rows where the conditions 's' in val or 'na' in val are true will have False in the resulting Series, and rows where the conditions are false will have True.

df[~...]: We use the boolean Series obtained from the apply() function as a mask to filter the rows in the DataFrame. By using ~ before the mask, we are selecting rows where the condition is True, which means rows that do not contain 's' or 'na' in any column.

So, after this code, df will contain only the rows where no column contains the letter 's' or the characters 'na', effectively removing rows with invalid or non-numeric values. This step is useful for data cleaning, ensuring that the remaining data is suitable for further analysis or processing.
